```python
# Question 2:
# Answer:

# Find Sum and Product of first 10 natural number using for loop.
sum=0
product=1
for i in range(1,11):
    sum = sum+i
    product = product*i
print("THE SUM OF FIRST 10 NATURAL NUMBER IS : ",sum)
print("THE PRODUCT OF FIRST 10 NATURAL NUMBER IS : ",product)

# Find Sum and Product of first 10 natural number using while loop.
sum1=0
product1=1
n=1
while n<=10:
    sum1 = sum1 + n
    product1 = product1 * n
    n = n + 1
print("THE SUM OF FIRST 10 NATURAL NUMBER IS : ",sum1)
print("THE PRODUCT OF FIRST 10 NATURAL NUMBER IS : ",product1)
    
    
```
